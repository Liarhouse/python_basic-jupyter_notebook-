```python
# 한 명의 학생에 대한 정보를 담아서 관리하는 클래스 선언
class Student:
    
    def __init__(self, hakbun, irum, kor, eng, math):
        self.hakbun = hakbun
        self.irum = irum
        self.kor = kor
        self.eng = eng
        self.math = math
    
    def getHakbun(self):
        return self.hakbun
    

    
    def getIrum(self):
        return self.irum
    
    def update(self, irum, kor, eng, math):
        self.irum = irum
        self.kor = kor
        self.eng = eng
        self.math = math
    
    def compute(self):
        self.tot = self.kor + self.eng + self.math
        self.avg = self.tot / 3
        
        if self.avg >= 90:
            self.grade = "A"
        elif self.avg >= 80:
            self.grade = "B"
        elif self.avg >= 70:
            self.grade = "C"
        elif self.avg >= 60:
            self.grade = "D"
        else:
            self.grade = "F"

    def disp(self):
        return "%s님의 총점은 %d이고, 평균은 %.2f이며, 학점은 %s입니다." % (
            self.name,self.tot, self.avg, self.grade)
    
    def show(self):
        return "%5s %5s %5d %.2f" % (self.hakbun, self.irum, self.tot, self.avg)
```

```python
class StuManager:
    
    def __init__(self):
        self.stuList = []
        
    def addStudent(self):
        hakbun = input("학번 = ")
        if hakbun in StuList:
            print("이미 존재하는 학번입니다.")
        else:
            irum = input("이름 = ")
            kor = int(input("국어 = "))
            eng = int(input("영어 = "))
            math = int(input("수학 = "))
            stu = Student(hakbun, irum, kor, eng, math)
            StuList.append(hakbun)
            self.stuList.append(stu)
            print("학생 등록이 완료되었습니다.")

    
    
    def allStudent(self):
        if len(self.stuList) != 0:
            for stu in self.stuList:
                stu.compute()
                print(stu.show())
            print("학생정보 조회가 완료되었습니다.")
        else:
            print("시스템에 등록된 학생이 없습니다.")
    
    def oneIrum(self):
        if len(self.stuList) != 0:
            irum = input("이름 = ")
            check = 0
            for stu in self.stuList:
                if irum == stu.getIrum():
                    stu.compute()
                    print(stu.show())
                    check = 1
            if check == 0:
                print("해당 이름의 학생은 시스템에 등록되어 있지 않습니다.")
        else:
            print("시스템에 등록된 학생이 없습니다.")
    
    def oneHakbun(self):
        if len(self.stuList) != 0:
            hakbun = input("학번 = ")
            for stu in self.stuList:
                if hakbun == stu.getHakbun():
                    stu.compute()
                    print(stu.show())
                else:
                    print("해당 학번의 학생은 시스템에 등록되어 있지 않습니다.")
        else:
            print("시스템에 등록된 학생이 없습니다.")
        
    def change(self):
        if len(self.stuList) != 0:
            hakbun = input("학번 = ")
            for stu in self.stuList:
                if hakbun == stu.getHakbun():
                    irum = input("이름 = ")
                    kor = int(input("국어 = "))
                    eng = int(input("영어 = "))
                    math = int(input("수학 = "))
                    stu.update(irum, kor, eng, math)
                    
                else:
                    print("해당 학번의 학생은 시스템에 등록되어 있지 않습니다.")
        else:
            print("시스템에 등록된 학생이 없습니다.")
            
    def showMenu(self):
        print("\n == 성적관리 프로그램 ==")
        print("1. 학생 등록")
        print("2. 학번 검색")
        print("3. 성적 수정")
        print("4. 이름 검색")
        print("5. 전체 출력")
        print("6. 작업 종료")
```

```python
def main():
    handle = StuManager()
    
    while True:
        handle.showMenu()
        print()
        menu = input("menu = ")
        
        if menu == "1":
            handle.addStudent()
        elif menu == "2":
            handle.oneHakbun()
        elif menu == "3":
            handle.change()
        elif menu == "4":
            handle.oneIrum()
        elif menu == "5":
            handle.allStudent()
        elif menu == "6":
            print("작업을 종료합니다.")
            break
        else:
            print("메뉴의 선택이 올바르지 않습니다.")
```
