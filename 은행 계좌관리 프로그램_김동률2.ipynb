# 은행 계좌관리 프로그램

```python
class Account:
    
    # 생성자 메서드
    def __init__(self, num = ' ', name = ' ', money = 0):
        self.num = num
        self.name = name
        self.money = money

                
    
    # 멤버 메서드               
    ## 입금관리
    def bank_in(self, money_in): 
        self.money += money_in
        
    ## 출금관리
    def bank_out(self):
        while True:
            money_out = int(input("출금금액 = "))
            if self.money >= money_out:
                self.money -= money_out
                break
            else:
                print("통장잔액이 부족합니다.")
                   
  
    ## 계좌번호 가져오기
    def get_num(self):
        return self.num

    ## 돈가져오기
    def get_money(self):
        return self.money

    
    
    ## 전체고객 잔액조회
    def disp(self):
        print(self.num, self.name, self.money, sep = "            ")
```

```python
accountList = []
bank = Account()

while True:
    print("== 계좌관리 프로그램 ==")
    print("1. 계좌개설")
    print("2. 입금처리")
    print("3. 출금처리")
    print("4. 전체고객 잔액조회")
    print("5. 프로그램 종료")
    print("====================")
    
    a = int(input("메뉴 = " ))
    print()
    # 계좌개설
    if a == 1:
        print("== 계좌개설 ==")
        num = input("계좌번호 = ")
        while bank in accountList:
            if num in bank.get_num():
                print("이미 존재하는 계좌번호입니다.")
            else:
                break
            num = input("계좌번호 = ")
        name = input("고객이름 = ")
        money = int(input("예금금액 = "))
        bank = Account(num, name, money)
        accountList.append(bank)
        print("-------------")
        print("계좌개설이 완료되었습니다.\n")
            
    # 입금처리    
    elif a == 2:
        print("== 입금처리 ==")
        while bank in accountList:
            num = input('계좌번호 = ')
            if num in bank.get_num():
                money_in = int(input("입금금액 = "))
                bank.bank_in(money_in)
                print("통장잔액 = ", bank.get_money())
                break
            else:
                print("존재하지 않는 계좌입니다.")
        print("-------------")
        print("입금이 완료되었습니다.\n")
     
    # 출금처리
    elif a == 3:
        print("== 출금처리 ==")
        while bank in accountList:
            num = input("계좌번호 = ")
            if num in bank.get_num():
                print("출금가능액 = ", bank.get_money())
                bank.bank_out()
                print("통장잔액 =", bank.get_money())
                break
            else:
                print("존재하지 않는 계좌입니다.")
        
        print("-------------")
        print("출금이 완료되었습니다.\n")
     
    # 전체고객 잔액조회
    elif a == 4:
        print("== 전체고객 잔액조회 ==")
        print("계좌번호", "고객이름", "예금잔액", sep = "      ")
        print("--------------------------------------------")
        for bank in accountList:
            bank.disp()
        print("--------------------------------------------")
        tot = 0
        for bank in accountList:
            tot += bank.get_money()
        print("총 보유예금액 : ", tot ,"\n")
    
    # 프로그램 종료
    elif a == 5:
        print("우리 은행을 이용해주셔서 감사합니다")
        break
    
    # 오류
    else:
        print("잘못 입력하셨습니다.\n")
```
